```python
S0_utilisation_rate = pd.read_csv("S0_utilisation_rate.csv")
S0_utilisation_rate.head()
```

```python
# Set style in Seaborn to plot
sns.set(rc={"figure.dpi":300})
sns.set_context("paper")  # change to notebook maybe? That is how it was in Chapter 4 first draft version 
sns.set_style("ticks")
```

```python
# Plot all sites (delete binwidth to get sns default)
g = sns.histplot(data=S0_utilisation_rate, x="Ut_rate", color="darkorange", binwidth=5) # kde=True
g.set_xticks(np.arange(0, 100, 10))  # looks better
g.set(xlabel= "Utilisation rate (%)", ylabel= "Number of institutions")
g.set(title="All sites")
plt.savefig("Fig_4.5a_ok.pdf", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Separate df by Institution type (Schools) (when I reimport it to plot)
s0_utrate_schools = S0_utilisation_rate[S0_utilisation_rate.Institution != 'Clinic']
s0_utrate_schools.head()
```

```python
# Plot Schools (with binwidth=5 same as others)
# THIS IS THE PLOT I'M USING IN PPT FOR NOW
g=sns.histplot(data=s0_utrate_schools, x="Ut_rate", color="firebrick", binwidth=5) # kde=True
g.set_xticks(np.arange(0, 100, 10))  # looks better
g.set(xlabel= "Utilisation rate (%)", ylabel= "Number of institutions")
g.set(title="Schools")
plt.savefig("Fig_4.5b_ok.pdf", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Separate df by Institution type (Clinics) (when I reimport it to plot)
s0_utrate_clinics = S0_utilisation_rate[S0_utilisation_rate.Institution != 'School']
s0_utrate_clinics.head()
```

```python
g=sns.histplot(data=s0_utrate_clinics, x="Ut_rate", color="forestgreen", binwidth=5) # kde=True
g.set_xticks(np.arange(0, 100, 10))  # looks better than extending axis to 100
g.set_yticks(np.arange(0, 5, 1))
g.set(xlabel= "Utilisation rate (%)", ylabel= "Number of institutions")
g.set(title="Clinics")
plt.savefig("Fig_4.5c_ok.pdf", dpi=300, bbox_inches="tight")
plt.show()
```
