```python
# Imports LCUE df of all sites in Scenario 0 to 4 (S0-S4)
lcue_s0_s4 = pd.read_csv("lcue_s0_s4.csv") 
lcue_s0_s4.head()
```

```python
# Set style in Seaborn to plot
sns.set(rc={"figure.dpi":300})
sns.set_context("paper")  # change to notebook maybe? That is how it was in Chapter 4 first draft version 
sns.set_style("whitegrid")
```

```python
# to separate df by Institution type (by Schools)
lcue_s0_s4_schools = lcue_s0_s4[lcue_s0_s4.Institution != 'Clinic']
lcue_s0_s4_schools
```

```python
# Plot schools LCUE
g = sns.catplot(
    data=lcue_s0_s4_schools, kind= "box", palette= "Set2")
g.set(xlabel= "Scenarios", ylabel= "LCUE ($/kWh)")
g.set(ylim=(0, 5)) # to get rid of outlier
g.set_xticklabels(["S0", "S1", "S2", "S3", "S4"])
g.set(title="Schools")
sns.despine(top=False, right=False)

plt.savefig("Fig_4.10a.pdf", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# to separate df by Institution type (by Clinics)
lcue_s0_s4_clinics = lcue_s0_s4[lcue_s0_s4.Institution != 'School']
lcue_s0_s4_clinics
```

```python
# Plot clinics LCUE
g = sns.catplot(
    data=lcue_s0_s4_clinics, kind= "box", palette= "Set2")
g.set(xlabel= "Scenarios", ylabel= "LCUE ($/kWh)")
g.set(ylim=(0, 5)) # to get rid of outlier
g.set_xticklabels(["S0", "S1", "S2", "S3", "S4"])
g.set(title="Clinics")
sns.despine(top=False, right=False)

plt.savefig("Fig_4.10b.pdf", dpi=300, bbox_inches="tight")
plt.show()
```
