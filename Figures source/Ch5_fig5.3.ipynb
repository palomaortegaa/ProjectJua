```python
# Code used to create Figure 5.3 in Ch. 5 of Emissions intensity of all 166 sites in S0-S4 (boxplot) 
# Created 5Sep23 - for thesis VF submission (after viva)
# Based on Jupyter notebook: lcue_s0-s4_plots; emissions_s0-s4_plots; Ch4_fig4.10_ab

import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np # maybe don't need this

# Move to path where csv are saved
path = "/Users/po416/Library/CloudStorage/OneDrive-ImperialCollegeLondon/PhDOct2018/Projects (12Mar20)/Project Jua/Modelling PJua analysis/Results_outputs"

cwd = os.getcwd()
cwd

# Change to path where csv file is saved
os.chdir(path) # or put the whole path
```

```python
# Imports LCUE df of all sites in Scenario 0 to 4 (S0-S4)
emi_s0_s4 = pd.read_csv("emi_s0_s4.csv") 
emi_s0_s4.head()
```

```python
# df with all the 166 institutions
emi_s0_s4.describe().round(2)
```

```python
# to separate df by Institution type (by Schools)
emi_s0_s4_schools = emi_s0_s4[emi_s0_s4.Institution != 'Clinic']
emi_s0_s4_schools
```

```python
# df of only schools (n=148)
emi_s0_s4_schools.describe().round(2)
```

```python
# Set style in Seaborn to plot
sns.set(rc={"figure.dpi":300})
sns.set_context("paper")  # change to notebook maybe? That is how it was in Chapter 4 first draft version 
sns.set_style("whitegrid")
```

```python
# Plot schools Emissions Intensity VF for final submission
g = sns.catplot(
    data=emi_s0_s4_schools, kind= "box", palette= "Set2")
g.set(xlabel= "Scenarios", ylabel= "Emissions intensity (gCO2eq/kWh)")
g.set(ylim=(0, 1600)) # to get rid of outlier
g.set_xticklabels(["S0", "S1", "S2", "S3", "S4"])
g.set(title="Schools")
sns.despine(top=False, right=False)

plt.savefig("Fig_5.3_schools.pdf", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# to separate df by Institution type (by Clinics)
emi_s0_s4_clinics = emi_s0_s4[emi_s0_s4.Institution != 'School']
emi_s0_s4_clinics
```

```python
# df of only clinics (n=18)
emi_s0_s4_clinics.describe().round(2)
```

```python
# Plot clinics Emissions Intensity VF for final submission
g = sns.catplot(
    data=emi_s0_s4_clinics, kind= "box", palette= "Set2")
g.set(xlabel= "Scenarios", ylabel= "Emissions intensity (gCO2eq/kWh)")
g.set(ylim=(0, 1600)) # to get rid of outlier
g.set_xticklabels(["S0", "S1", "S2", "S3", "S4"])
g.set(title="Clinics")
sns.despine(top=False, right=False)

plt.savefig("Fig_5.3_clinics.pdf", dpi=300, bbox_inches="tight")
plt.show()
```
