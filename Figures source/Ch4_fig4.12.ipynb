```python
# Import csv of clinics to make plots
avg_dumped_energy_clinics = pd.read_csv("avg_dumped_energy_clinics.csv")
avg_dumped_energy_clinics.head().round(2)
```

```python
# Import sns for plotting and set style
import seaborn as sns

# Set style in Seaborn to plot
sns.set(rc={"figure.dpi":300})
sns.set_context("paper")  # change to notebook maybe? That is how it was in Chapter 4 first draft version 
sns.set_style("ticks")
```

```python
# Plot it as boxenplot (only clinics)
g = sns.boxenplot(data = avg_dumped_energy_clinics)
g.set(xlabel= "Hour of day", ylabel="Unutilised energy (kWh)")
g.set_title("S0 - Clinics")
plt.savefig("Fig_4.12a.pdf", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Import csv of schools to make plots
avg_dumped_energy_schools = pd.read_csv("avg_dumped_energy_schools.csv")
avg_dumped_energy_schools.head().round(2)
```

```python
# Plot it as boxenplot (only schools)
g = sns.boxenplot(data = avg_dumped_energy_schools)
#g.set(ylim=(-0.05, 1.4)) # extend y-limit to compare it with clinics
g.set(xlabel= "Hour of day", ylabel="Unutilised energy (kWh)")
g.set_title("S0 - Schools")
#plt.savefig("Fig_4.12a.pdf", dpi=300, bbox_inches="tight")
plt.show()
```
